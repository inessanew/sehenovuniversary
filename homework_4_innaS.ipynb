```python
#task12 - Fibonacci sequence #создано с помощью подсказок Colab
n = int(input('Введите число: '))

if n <= 0:
    print('Пожалуйста, введите положительное натуральное число.')
elif n == 1:
    print(0)
else:
    a, b = 0, 1
    fib_sequence = []
    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b
    print(' '.join(map(str, fib_sequence)))
```
